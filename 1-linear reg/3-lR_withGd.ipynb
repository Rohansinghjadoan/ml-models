```python
class LinearRegressionGD:
    def __init__(self, learning_rate=0.01, epochs=1000):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.m = 0.0
        self.b = 0.0

    def fit(self, x, y):
        n = len(x)

        for _ in range(self.epochs):
            dm = 0
            db = 0

            for i in range(n):
                y_pred = self.m * x[i] + self.b
                error = y[i] - y_pred

                dm += -2 * x[i] * error
                db += -2 * error

            dm /= n
            db /= n

            self.m -= self.learning_rate * dm
            self.b -= self.learning_rate * db

    def predict(self, x):
        prediction = []
        for xi in x:
            prediction.append(self.m * xi + self.b)
        return prediction
```

```python

X_train=[1,2,3,4,5]
y_train=[2,4,6,8,10]
```

```python

model=LinearRegressionGD(learning_rate=0.01,epochs=1000)
```

```python
model.fit(X_train,y_train)
```

```python
X_test=[6,8]
prediction=model.predict(X_test)
print(prediction)
```
