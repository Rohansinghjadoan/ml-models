```python
import numpy as np

class LinearSVM:
    def __init__(self, learning_rate=0.001, epochs=1000, C=1.0):
        self.lr = learning_rate
        self.epochs = epochs
        self.C = C
        self.w = None
        self.b = None

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Initialize parameters
        self.w = np.zeros(n_features)
        self.b = 0

        for epoch in range(self.epochs):
            for i in range(n_samples):
                condition = y[i] * (np.dot(X[i], self.w) + self.b)

                if condition >= 1:
                    # Only regularization term
                    self.w -= self.lr * self.w
                else:
                    # Regularization + hinge loss
                    self.w -= self.lr * (self.w - self.C * y[i] * X[i])
                    self.b -= self.lr * (-self.C * y[i])

    def predict(self, X):
        decision = np.dot(X, self.w) + self.b
        return np.sign(decision)
```
