```python
import math
class LogisticReg():
    def __init__(self,learning_rate=0.01,epochs=1000):
        self.learning_rate=learning_rate
        self.epochs=epochs
        self.m=0
        self.b=0
    def fit(self,x,y):
        n=len(x)
        for epoch in range(self.epochs):
            dm=0
            db=0
            for i in range(n):
                z=self.m*x[i]+self.b
                y_pred=self.sigmoid(z)
                error=y_pred-y[i]
                dm+=error*x[i]
                db+=error
            dm/=n
            db/=n
            self.m-=self.learning_rate*dm
            self.b-=self.learning_rate*db
    def predProba(self,x):
        probablities=[]
        for i in x:
            z=self.m*i+self.b
            probablities.append(self.sigmoid(z))
        return probablities
    def prediction(self,x,thershold=0.5):
        prediction=[]
        for p in self.predProba(x):
            prediction.append(1 if p>thershold else 0)
        return prediction


            
    def sigmoid(self,x):
        return 1/(1+math.exp(-x))



```
